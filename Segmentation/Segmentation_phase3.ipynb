```python
import os
import shutil
import re

# Input
input_dir = "/content/drive/MyDrive/be5370_hw1/hw3/patches_for_training_20251003"
# Output
dataset_dir = "/content/drive/MyDrive/be5370_hw1/colab_data/nnUNet_raw/Dataset101_BrainSeg"

imagesTr = os.path.join(dataset_dir, "imagesTr")
labelsTr = os.path.join(dataset_dir, "labelsTr")
os.makedirs(imagesTr, exist_ok=True)
os.makedirs(labelsTr, exist_ok=True)

pattern = re.compile(r"(subj\d+)_team\d+_(patch\d+)_(.+)\.nii\.gz")

for fname in os.listdir(input_dir):
    if not fname.endswith(".nii.gz"):
        continue
    match = pattern.match(fname)
    if not match:
        continue

    subj, patch, modality = match.groups()
    case_id = f"{subj}_{patch}"
    src = os.path.join(input_dir, fname)

    if "ciss" in modality.lower():
        dest = os.path.join(imagesTr, f"{case_id}_0000.nii.gz")
    elif "t2" in modality.lower():
        dest = os.path.join(imagesTr, f"{case_id}_0001.nii.gz")
    elif "final" in modality.lower():
        dest = os.path.join(labelsTr, f"{case_id}.nii.gz")
    else:
        continue

    print(f"Copying {fname} -> {dest}")
    shutil.copy(src, dest)

print("Reorganization complete.")
```

# STEP 3 - E

To improve the quality of nnU-Net segmentations and remove spurious foreground labels in non-tissue regions, I implemented a post-processing pipeline that leverages both connected component analysis and tissue masking. For each segmentation, I first generated a binary tissue mask from the corresponding raw CISS image using SimpleITK's Otsu thresholding. This mask identifies regions likely to contain brain tissue. I then used sitk.ConnectedComponent to label all contiguous foreground regions in the segmentation. For each connected component, I calculated its size and the degree of overlap with the tissue mask. Only components larger than a minimum size (1000 voxels) and with at least 30% overlap with the tissue mask were retained; all others were set to background.

```python
import os
import SimpleITK as sitk
import numpy as np

def generate_tissue_mask(image):
    otsu_filter = sitk.OtsuThresholdImageFilter()
    otsu_filter.SetInsideValue(0)
    otsu_filter.SetOutsideValue(1)
    mask = otsu_filter.Execute(image)
    return mask

def postprocess_segmentation(segmentation, otsu_mask=None, min_size=1000):
    cc = sitk.ConnectedComponent(segmentation > 0)
    stats = sitk.LabelShapeStatisticsImageFilter()
    stats.Execute(cc)

    cc_array = sitk.GetArrayFromImage(cc)
    seg_array = sitk.GetArrayFromImage(segmentation)
    cleaned_array = np.zeros(seg_array.shape, dtype=seg_array.dtype)

    if otsu_mask:
        otsu_array = sitk.GetArrayFromImage(otsu_mask)
        for label in stats.GetLabels():
            size = stats.GetNumberOfPixels(label)
            if size < min_size:
                continue

            component_mask = (cc_array == label)
            overlap = np.sum(component_mask & (otsu_array > 0))
            overlap_ratio = overlap / size

            if overlap_ratio > 0.3:
                cleaned_array[component_mask] = seg_array[component_mask]
    else:
        center = np.array(seg_array.shape) / 2
        for label in stats.GetLabels():
            size = stats.GetNumberOfPixels(label)
            if size < min_size:
                continue

            component_mask = (cc_array == label)
            indices = np.argwhere(component_mask)
            mean_pos = indices.mean(axis=0)
            if np.linalg.norm(mean_pos - center) < min(seg_array.shape) / 3:
                cleaned_array[component_mask] = seg_array[component_mask]

    post_processed_seg = sitk.GetImageFromArray(cleaned_array)
    post_processed_seg.CopyInformation(segmentation)

    return post_processed_seg


raw_folder = "/content/drive/MyDrive/be5370_hw1/hw3/hemi_testing"
seg_folder = "/content/drive/MyDrive/be5370_hw1/hw3/hemi_testing_nnUNet_output"
output_folder = "/content/drive/MyDrive/be5370_hw1/hw3/output_stepE/"

os.makedirs(output_folder, exist_ok=True)

seg_files = [f for f in os.listdir(seg_folder) if f.endswith(".nii.gz")]

for seg_file in seg_files:
    seg_path = os.path.join(seg_folder, seg_file)
    raw_file = seg_file.split('.')[0] + '_ciss.nii.gz'
    raw_path = os.path.join(raw_folder, raw_file)

    segmentation = sitk.ReadImage(seg_path)
    raw_image = sitk.ReadImage(raw_path)

    otsu_mask = generate_tissue_mask(raw_image)

    post_processed_seg = postprocess_segmentation(segmentation, otsu_mask, min_size=1000)

    output_path = os.path.join(output_folder, f"postprocessed_{seg_file}")
    sitk.WriteImage(post_processed_seg, output_path)

    print(f"Processed and saved: {output_path}")
```

# Step III-F



The Medial Frontal Cortex had a postive correlation at r = 0.65 and p = 0.056 while the Motor Cortex and Anterior Temporal Pole had negative close to zero correlations. Based on the plot, we can see that there is a substantial deviation from the y=x line which indicates that the agreement is not ideal. Therefore, the segmentation is not high-quality.

```python
import os
import re
import glob
import math
import pandas as pd
import numpy as np
import SimpleITK as sitk
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from upenn_be5370_utils.landmarks import hemi_landmark_thickness

test_folder = "/content/drive/MyDrive/be5370_hw1/hw3/test2"
prediction_folder = "/content/drive/MyDrive/be5370_hw1/hw3/predictions"

csv_columns = {
    "layout": "wide",
    "subject": "Id",
    "motor": "MotorCortex",
    "medfr": "MedialFrontalCortex",
    "anttemp": "AnteriorTemporalPole"
}

def canonicalize_subject(x):
    s = str(x)
    s = re.sub(r"\.nii\.gz$","",s)
    m = re.search(r"(subj_\d+)", s)
    if m: return m.group(1)
    if s.isdigit(): return f"subj_{int(s):03d}"
    m = re.search(r"(\d{2,3})", s)
    if m: return f"subj_{int(m.group(1))%1000:03d}"
    return s

subs = sorted(glob.glob(os.path.join(prediction_folder, "subj_*.nii.gz")))
rows = []
for seg_path in subs:
    subj = re.search(r"(subj_\d+)", os.path.basename(seg_path)).group(1)
    dots_path = os.path.join(test_folder, f"{subj}_dots.nii.gz")
    if not os.path.exists(dots_path):
        print(f"Missing dots file for {subj}")
        continue

    th = hemi_landmark_thickness(seg_path, dots_path)

    for lm in ["MotorCortex","MedialFrontalCortex","AnteriorTemporalPole"]:
        v = th.get(lm)
        if v is not None:
            rows.append({"subject": subj, "landmark": lm, "predicted": float(v)})

pred_df = pd.DataFrame(rows)
csv_path = os.path.join(test_folder, "reference_thickness.csv")

df = pd.read_csv(csv_path)
print("CSV loaded with columns:", list(df.columns))

ref_df = []
for _, row in df.iterrows():
    subj = canonicalize_subject(row[csv_columns["subject"]])
    ref_df.append({"subject": subj, "landmark": "MotorCortex", "reference": row[csv_columns["motor"]]})
    ref_df.append({"subject": subj, "landmark": "MedialFrontalCortex", "reference": row[csv_columns["medfr"]]})
    ref_df.append({"subject": subj, "landmark": "AnteriorTemporalPole", "reference": row[csv_columns["anttemp"]]})
ref_df = pd.DataFrame(ref_df)


merged = pd.merge(pred_df, ref_df, on=["subject","landmark"], how="inner")
corrs = []
for lm, g in merged.groupby("landmark"):
    r, p = pearsonr(g["reference"], g["predicted"])
    corrs.append({"landmark": lm, "r": r, "p": p, "n": len(g)})
corrs = pd.DataFrame(corrs)
print("Pearson correlation by landmark")
print(corrs.to_string(index=False))

plt.figure(figsize=(7,7))
for lm, g in merged.groupby("landmark"):
    plt.scatter(g["reference"], g["predicted"], label=f"{lm}", s=50)
mn, mx = merged[["reference","predicted"]].min().min(), merged[["reference","predicted"]].max().max()
plt.plot([mn,mx],[mn,mx],"--",c="gray")
plt.xlabel("Reference Thickness")
plt.ylabel("nnU-Net Thickness")
plt.title("nnU-Net vs Reference")
plt.legend(); plt.grid(alpha=0.3); plt.tight_layout(); plt.show()

merged.to_csv(os.path.join(prediction_folder,"landmark_thickness_comparison.csv"),index=False)
print("\nSaved merged results.")
```
